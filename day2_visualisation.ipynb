```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
!curl -L http://bit.ly/dw_car_data -o car.h5
```

```
df = pd.read_hdf('car.h5')
df.shape
```

```
df.columns.values
```

```
df['param_year'].plot(kind='bar', figsize=(15, 5))
```

```
(df.groupby('param_year')['param_marka-pojazdu']
 .agg(np.size)
 ).plot(kind='bar', figsize=(18, 4))
```

```
(df.groupby('param_year')['seller_type'=='owner']
 .agg(np.size)
 ).plot(kind='bar', figsize=(18, 4))
```

```
df.groupby('param_year').count()
```

```
df.groupby('param_marka-pojazdu')['price_value'].mean()
```

```
(df.groupby('param_marka-pojazdu')['price_value']
 .agg([np.mean, np.median, np.size])
 .sort_values(by='mean')
 ).plot(kind='bar', figsize=(15, 5))
```

```
(df.groupby('param_marka-pojazdu')['price_value']
 .agg([np.mean, np.median, np.size])
 .sort_values(by='mean')
 .head(50)
 ).plot(kind='bar', figsize=(15, 5), subplots=True)
```

```
(df.['param_year']
 ).plot(kind='bar', figsize=(15, 5))
```
